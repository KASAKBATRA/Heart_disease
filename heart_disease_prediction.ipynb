```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart_disease_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace= True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Smoking'] = le.fit_transform(df['Smoking'])
df['Alcohol Intake'] = le.fit_transform(df['Alcohol Intake'])
df['Chest Pain Type'] = le.fit_transform(df['Chest Pain Type'])
df['Obesity'] = le.fit_transform(df['Obesity'])
df['Family History'] = le.fit_transform(df['Family History'])
df['Heart Disease'] = le.fit_transform(df['Heart Disease'])
df['Exercise Induced Angina'] = le.fit_transform(df['Exercise Induced Angina'])
df['Diabetes'] = le.fit_transform(df['Diabetes'])
df
```

```python
df.info()
```

```python
df.corr()
```

```python
df.drop('Exercise Hours', axis = 1, inplace = True)
df.drop('Obesity', axis = 1, inplace = True)
df.drop('Blood Sugar', axis = 1, inplace = True)
df.drop('Smoking', axis= 1, inplace = True)
df.drop('Chest Pain Type', axis= 1, inplace = True)
df.drop('Family History', axis = 1, inplace = True)
df.drop('Stress Level', axis = 1, inplace = True)
```

```python
df
```

```python
df.corr()
```

```python
X = df.iloc[:,:-1].values
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
#spiltting

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y , test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# **LOGISTIC**

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
accuracy_score(Y_test, Y_pred)
```

# **KNN**

```python
X = df.iloc[:,:-1].values
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
#spiltting

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y , test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# model training

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 11, metric = 'minkowski')
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
#confusion matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
accuracy_score(Y_test, Y_pred)
```

# **SUPPORT VECTOR MACHINE**

```python
Y_pred = classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```
